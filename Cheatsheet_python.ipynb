## Formatting and Regex

### Two types of formatting are available in Python2 and 3
#### 1. Old style(%)
#### 2. New style({})

### 1. old style:
- %s string
- %d decimal integer
- %x hex integer
- %o octal integer
- %f decimal float
- %e exponential float
- %g decimal or exponential float %% a literal%
#### Ex. 1
- actor = 'Richard Gere'
- cat = 'Chester'
- weight = 28
- "Our cat %s weighs %s pounds" % (cat, weight)
#### Ex. 2
- n=42 
- f=7.03
- s = 'string cheese'
- --> '%d%f%s'%(n,f,s) --> Format using default width
- --> '%10d %10f %10s' % (n, f, s) --> Set a minimum field width of 10 characters for each variable, and align them to the right, 
- --> '%-10d %-10f %-10s' % (n, f, s) --> set same width as above with left align
- --> '%10.4d %10.4f %10.4s' % (n, f, s) --> same field width, but a maximum character width of 4,
- --> '%.4d %.4f %.4s' % (n, f, s) --> same song as before, but right-aligned
- --> '%*.*d %*.*f %*.*s' % (10, 4, n, 10, 4, f, 10, 4, s) --> get the field widths from arguments rather than hard-coding them

### New Style Formatting
- '{} {} {}'.format(n, f, s) --> 
- '{2} {0} {1}'.format(f, s, n) --> specify the order of arguments
- '{n} {f} {s}'.format(n=42, f=7.03, s='string cheese') --> The arguments can be a dictionary or named arguments, and the specifiers can include their names
- d = {'n': 42, 'f': 7.03, 's': 'string cheese'} -->combining our three values into a dictionary
- --> '{0[n]} {0[f]} {0[s]} {1}'.format(d, 'other') --> {0} is the entire dictionary, whereas {1} is the string 'other' that follows the dictionary
- '{0:d} {1:f} {2:s}'.format(n, f, s) --> printed their arguments with default formats as old styles
- '{n:d} {f:f} {s:s}'.format(n=42, f=7.03, s='string cheese') --> use the same values, but as named arguments
- '{0:10d} {1:10f} {2:10s}'.format(n, f, s) --> Minimum field width 10, right-aligned (default)
- '{0:>10d} {1:>10f} {2:>10s}'.format(n,f,s) -> Same as the preceding example,but the > char. make the right-alignment
- '{0:<10d} {1:<10f} {2:<10s}'.format(n, f, s) -> Minimum field width 10, left-aligned
- '{0:^10d} {1:^10f} {2:^10s}'.format(n, f, s) - Minimum field width 10, centered
#### There is one change from old-style: the precision value (after the decimal point) still means the number of digits after the decimal for floats, and the maximum number of characters for strings, but you canâ€™t use it for integers
- '{0:>10.4d} {1:>10.4f} {2:10.4s}'.format(n, f, s) -> upto decimal value
- '{0:!^20s}'.format('BIG SALE')--> fill character. If you want something other than spaces to pad your output fields, put it right after the :, before any alignment (<, >, ^) or width specifiers
## Regular Expressions
- result = re.match('You', 'Young Frankenstein') -> Here, 'You' is the pattern and 'Young Frankenstein' is the source
- youpattern = re.compile('You') -> For more complex matches,we can compile pattern first to speed up the match later
- --> result = youpattern.match('Young Frankenstein')
### Options to use in RE
â€¢ match() is not the only way to compare the pattern and source. Here are several other methods you can use:
â€¢ search() returns the first match, if any.
â€¢ findall() returns a list of all non-overlapping matches, if any.
â€¢ split() splits source at matches with pattern and returns a list of the string pieces.
â€¢ sub() takes another replacement argument, and changes all parts of source that are matched by pattern to replacement.
#### Exact match with match()
- --> import re
- --> source = 'Young Frankenstein'
- --> m = re.match('You', source) # match starts at the beginning of source 
- --> if m: # match returns an object; do this to see what matched
    - --> print(m.group())
- m = re.match('^You', source) # start anchor does the same
- if m:
    - print(m.group())
- m = re.match('Frank', source)
- if m:
    - print(m.group())
